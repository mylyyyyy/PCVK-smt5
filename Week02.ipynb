###5.Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!

```python
import random
width, height = 400, 400
katkat = kat.copy()
for _ in range(5):
    x1, y1 = random.randint(50, width - 50), random.randint(50, height - 50)
    x2, y2 = x1 + random.randint(20, 100), y1 + random.randint(20, 100)
    color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))
    cv.rectangle(katkat, (x1, y1), (x2, y2), color, -1)  # Kotak diisi dengan warna

plt.imshow(katkat)
```
